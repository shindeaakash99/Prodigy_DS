```python
#Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, 
#such as the distribution of ages or genders in a population
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv('train.csv')
```

```python
# Distribution of passengers by Class on the Titanic
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Pclass', palette='viridis')
plt.title('Distribution of Passengers by Class on the Titanic')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
#Distribution of Passenger Ages (Histogram)
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='Age', bins=30, kde=True, color='blue')
plt.title('Distribution of Passenger Ages on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#Distribution of Passenger Genders (Bar Chart)
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Sex', palette='Set2')
plt.title('Distribution of Passengers by Gender on the Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Survival Rate by Passenger Class (Bar Chart)
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Pclass', y='Survived', palette='Set1')
plt.title('Survival Rate by Passenger Class on the Titanic')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
#Distribution of Fares (Histogram)
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='Fare', bins=40, kde=True, color='green')
plt.title('Distribution of Fares on the Titanic')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```
